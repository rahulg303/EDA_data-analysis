```python
## Feature Engeneering
# Extracting day, month and year and creating new columns from the Date_of_Journey column
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Date']
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
# Date of Journey
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
df.head(1)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
df.head()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Arrival Time
df['Arr_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arr_Minute']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arr_Hour']=df['Arr_Hour'].astype(int)
df['Arr_Minute']=df['Arr_Minute'].astype(int)
df.info()
```

```python
df.drop(['Arrival_Time'], axis=1, inplace=True)
df.head()
```

```python
# Departure time
df['Dep_hr']=df['Dep_Time'].str.split(':').str[0]
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
df.head(3)
```

```python
df['Dep_hr']=df['Dep_hr'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
df.info()
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)
df.head(2)
```

```python
df['Route'].value_counts()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
df.head()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df.drop(['Route'], axis=1, inplace=True)
df.head()
```
